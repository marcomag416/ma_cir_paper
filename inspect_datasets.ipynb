## MSCOCO

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
from custom_datasets.mscoco import build_mscoco_dataset

mscoco_val = build_mscoco_dataset(
    split="val",
    image_transform=None,
    caption_transform=None,
)

def view_sample(index):
    sample = mscoco_val[index]
    image = sample['pixel_values']
    caption = sample['input_ids'] if sample['input_ids'] is not None else "No caption available"
 
    clear_output(wait=True)
    print(f"Index: {index}, Caption: {caption}")
    display(image)

widgets.interact(view_sample, index=widgets.IntSlider(min=0, max=len(mscoco_val)-1, value=0));
```

## CIRR

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
from custom_datasets.cirr import build_cirr_dataset

cirr_val = build_cirr_dataset(
    split="val",
    image_transform=None,
    caption_transform=None,
    mode="triplets"
)

def view_sample(index):
    clear_output(wait=True)
    sample = cirr_val[index]
    
    caption = sample['transformed_caption'] if sample['transformed_caption'] is not None else "No caption available"
    print(f"Index: {index}, Caption: {caption}")

    reference = sample['reference']
    target = sample['target']
    display(reference)
    display(target)

widgets.interact(view_sample, index=widgets.IntSlider(min=0, max=len(cirr_val)-1, value=0));
```
